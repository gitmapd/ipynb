```python
import requests
```

```python
urls=[]
```

```python
def scrape(site):
    r=requests.get(site)
    s=BeautifulSoup(r.text,"html.parser")
    for i in s.find_all("a"):
        print(i.text)
        links=i.get('href')
        urls.append(links)
```

```python
site="https://omega1239.webnode.pt/2021-2021/"
scrape(site)
```
